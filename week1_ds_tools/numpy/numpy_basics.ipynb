Covers: array creation, indexing/slicing, broadcasting, vectorised ops, linear algebra.

```python
import numpy as np
print('NumPy version:', np.__version__)
```

## 1. Array Creation

```python
# 1D, 2D arrays
a = np.array([1, 2, 3, 4, 5])
b = np.array([[1, 2, 3], [4, 5, 6]])

# Useful constructors
zeros   = np.zeros((3, 3))
ones    = np.ones((2, 4))
eye     = np.eye(3)
rng     = np.arange(0, 10, 2)      # [0, 2, 4, 6, 8]
linsp   = np.linspace(0, 1, 5)     # 5 evenly spaced points
rand    = np.random.randn(3, 3)    # Standard normal

print(b.shape, b.dtype)
```

## 2. Indexing & Slicing

```python
arr = np.arange(12).reshape(3, 4)
print(arr)

print(arr[1, 2])        # Single element
print(arr[:, 1])        # Whole column
print(arr[0:2, 1:3])   # Sub-matrix
print(arr[arr > 5])     # Boolean mask
```

## 3. Broadcasting

```python
# Broadcasting: smaller array stretched to match larger
x = np.array([[1, 2, 3], [4, 5, 6]])   # (2, 3)
row_bias = np.array([10, 20, 30])       # (3,) -> broadcast to (2, 3)

print(x + row_bias)
print(x * 2)            # Scalar broadcast
```

## 4. Vectorised Operations & Aggregations

```python
data = np.random.randn(100)

print('Mean  :', data.mean())
print('Std   :', data.std())
print('Min   :', data.min())
print('Max   :', data.max())
print('Sum   :', data.sum())

# Axis-wise aggregation
m = np.random.randint(1, 10, size=(4, 5))
print('Row sums   :', m.sum(axis=1))
print('Col means  :', m.mean(axis=0))
```

## 5. Linear Algebra

```python
A = np.array([[2, 1], [5, 3]])
B = np.array([[1, 0], [0, 1]])

print('Dot product:\n', A @ B)              # Matrix multiply
print('Transpose:\n', A.T)
print('Determinant:', np.linalg.det(A))
print('Inverse:\n', np.linalg.inv(A))

# Solving Ax = b
b_vec = np.array([4, 7])
x = np.linalg.solve(A, b_vec)
print('Solution x:', x)
```
