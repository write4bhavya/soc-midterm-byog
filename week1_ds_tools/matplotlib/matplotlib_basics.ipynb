Covers: line/scatter/bar/histogram plots, subplots, styling, saving figures.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

## 1. Line Plot

```python
x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x), label='sin(x)', color='steelblue', lw=2)
ax.plot(x, np.cos(x), label='cos(x)', color='tomato',    lw=2, linestyle='--')
ax.set_title('Trigonometric Functions')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

## 2. Scatter Plot

```python
np.random.seed(42)
x_s = np.random.randn(100)
y_s = 2 * x_s + np.random.randn(100) * 0.5

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(x_s, y_s, alpha=0.6, edgecolors='k', linewidths=0.4)
ax.set_title('Scatter: linear relationship')
ax.set_xlabel('X'); ax.set_ylabel('Y')
plt.tight_layout(); plt.show()
```

## 3. Bar Chart & Histogram

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar
categories = ['Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5']
values     = [3, 6, 5, 4, 2]
axes[0].bar(categories, values, color='mediumslateblue', edgecolor='k', linewidth=0.6)
axes[0].set_title('Notebooks per Week')

# Histogram
data = np.random.randn(1000)
axes[1].hist(data, bins=30, color='teal', edgecolor='white', linewidth=0.5)
axes[1].set_title('Normal Distribution')

plt.tight_layout(); plt.show()
```

## 4. Subplots Grid & Saving

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
x = np.linspace(-3, 3, 300)

funcs = [np.sin, np.cos, np.tanh, lambda t: t**2 - 1]
names = ['sin', 'cos', 'tanh', 'x² - 1']
colors = ['steelblue', 'tomato', 'seagreen', 'darkorange']

for ax, f, name, c in zip(axes.flat, funcs, names, colors):
    ax.plot(x, f(x), color=c, lw=1.8)
    ax.set_title(name)
    ax.grid(alpha=0.3)

fig.suptitle('Common Functions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../../assets/functions_grid.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to assets/functions_grid.png')
```
