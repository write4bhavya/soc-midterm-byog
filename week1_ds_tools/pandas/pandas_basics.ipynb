```python
import pandas as pd
import numpy as np
print('Pandas version:', pd.__version__)
```

## 1. Series & DataFrame

```python
# Series
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)

# DataFrame from dict
df = pd.DataFrame({
    'Name'  : ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age'   : [24, 27, 22, 29],
    'Score' : [88.5, 72.0, 91.3, 65.8]
})
df
```

## 2. Selection & Filtering

```python
print(df['Name'])                    # Column
print(df[['Name', 'Score']])         # Multiple columns
print(df.loc[0:1])                   # Rows by label
print(df.iloc[2:4])                  # Rows by position
print(df[df['Score'] > 80])          # Boolean filter
print(df.query('Age < 27 and Score > 70'))  # Query syntax
```

## 3. Data Cleaning

```python
messy = pd.DataFrame({
    'A': [1, None, 3, None],
    'B': ['x', 'y', None, 'z'],
    'C': [1.1, 2.2, 2.2, 3.3]
})

print(messy.isnull().sum())          # Missing count per col
print(messy.fillna({'A': 0, 'B': 'unknown'}))  # Fill
print(messy.dropna())                # Drop rows with any NaN
print(messy.drop_duplicates(subset='C'))  # Deduplicate
```

## 4. GroupBy & Aggregation

```python
sales = pd.DataFrame({
    'Region'  : ['North', 'South', 'North', 'East', 'South'],
    'Product' : ['A', 'B', 'A', 'C', 'A'],
    'Revenue' : [100, 200, 150, 300, 250]
})

print(sales.groupby('Region')['Revenue'].sum())
print(sales.groupby(['Region', 'Product']).agg({'Revenue': ['sum', 'mean']}))
```

## 5. Merging & Joining

```python
left  = pd.DataFrame({'id': [1,2,3], 'name': ['A','B','C']})
right = pd.DataFrame({'id': [2,3,4], 'score': [90, 85, 70]})

print(pd.merge(left, right, on='id', how='inner'))   # Inner join
print(pd.merge(left, right, on='id', how='left'))    # Left join
```

## 6. Time-Series

```python
dates = pd.date_range('2024-01-01', periods=60, freq='D')
ts = pd.Series(np.random.randn(60).cumsum(), index=dates)

print(ts.resample('W').mean())       # Weekly averages
print(ts.rolling(window=7).mean())   # 7-day rolling mean
print(ts['2024-02'])                 # Slice by month
```
